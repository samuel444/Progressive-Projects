```python
import sys

import pandas as pd
import matplotlib.pyplot as plt

sys.path.append("/Users/sam/Progressive-Projects/Packages")

import my_packages as mp
```

# 1. Create the Dataset

Creating a Dataframe just like you would in a normal python file, however it gets displayed cleaner in Notebooks!

```python
df = pd.DataFrame({
    "Day": [1,2,3,4,5,6,7],
    "Price": [100,102,101,105,107,106,110]
})

df
```

# 2. Changing DataFrames

Rather than modifying the original data directly, we create new columns that may be useful during analysis.

Here we'll calculate:

- Daily Returns
- Z-Scores

```python
df["Return"] = df["Price"].pct_change()

df["Z Score"] = mp.z_score(df["Price"])

df
```

# 3. Visualising the Data

Visualising the Data with Matplotlib

```python

plt.plot(df["Day"], df["Price"])

plt.title("Price Over Time")
plt.xlabel("Day")
plt.ylabel("Price")

plt.grid(True)

plt.show()
```
