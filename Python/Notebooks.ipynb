```python
import sys

import pandas as pd
import matplotlib.pyplot as plt

sys.path.append("/Users/sam/Progressive-Projects/Packages")

import my_packages as mp
```

# 1. Create the Dataset

Normally this data would come from a CSV, database or API.

For learning purposes we'll create a very small DataFrame manually.

```python
df = pd.DataFrame({
    "Day": [1,2,3,4,5,6,7],
    "Price": [100,102,101,105,107,106,110]
})

df
```

# 2. Feature Engineering

Rather than modifying the original data directly, we create new columns that may be useful during analysis.

Here we'll calculate:

- Daily Returns
- Z-Scores

```python
df["Return"] = df["Price"].pct_change()

df["Z Score"] = mp.z_score(df["Price"])

df
```

# 3. Summary Statistics

A quick statistical overview often provides useful context before visualising the data.

```python
df.describe()
```

# 4. Visualising the Data

Visualisations often reveal trends that are difficult to identify from raw numbers alone.

```python
plt.figure(figsize=(8,4))

plt.plot(df["Day"], df["Price"])

plt.title("Price Over Time")
plt.xlabel("Day")
plt.ylabel("Price")

plt.grid(True)

plt.show()
```
